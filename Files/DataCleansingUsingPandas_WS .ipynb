DATA CLEANSING using PANDAS LIBRARY

```python
#Importing Pandas Library
import pandas as pd 

#Path of Data
path="https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data" 

df=pd.read_csv(path,header=None)

headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
"drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
"num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
"peak-rpm","city-mpg","highway-mpg","price"]

df.columns=headers
df.head(10)
```

```python
#importing NumPy Library
import numpy as np

#replace "?" to NaN
df.replace("?", np.nan, inplace=True) #inPlace=True makes the changes permanent in the Dataset
df.head(10)
```

```python
missing_data=df.isnull() #to check if any data set is NULL or NOT
missing_data.head(5)
```

```python
missing_data=df.notnull() #to check if any data set is NOT-NULL or NOT
missing_data.head(5)
```

```python
for column in missing_data.columns.values.tolist():
    print(column)
    print(missing_data[column].value_counts())
    print(" ")
```

```python
avg_1=df["normalized-losses"].astype("float").mean(axis=0)
avg_1
```

```python
df["normalized-losses"].replace(np.nan, avg_1, inplace= True)
df.head(5)
```

```python
avg_2=df['bore'].astype('float').mean(axis=0)
avg_2
```

```python
df['bore'].replace(np.nan,avg_2, inplace=True)
df.head(5)
```

```python
avg_3=df['stroke'].astype('float').mean(axis=0)
avg_3
```

```python
df['stroke'].replace(np.nan,avg_3, inplace=True)
df.head(5)
```
